```python
%sql
WITH tb_pedidos AS (
  SELECT *
  FROM workspace.olist.orders
  WHERE order_purchase_timestamp < '2018-07-01'
)
--SELECT * FROM tb_pedidos;

select toi.seller_id, 
       round(avg(case when tor.review_score = 1 then 1 else 0 end),2) as pct_nota1,
       round(avg(case when tor.review_score = 2 then 1 else 0 end),2) as pct_nota2,

       round(avg(case when (date_diff(tp.order_purchase_timestamp, '2018-07-01') <= 28) and (tor.review_score = 1) then 1 else 0 end),2) as pct_nota1_d28,
       round(avg(case when (date_diff(tp.order_purchase_timestamp, '2018-07-01') <= 28) and (tor.review_score = 2) then 1 else 0 end),2) as pct_nota2_d28,

       round(avg(case when (date_diff(tp.order_purchase_timestamp, '2018-07-01') <= 56) and (tor.review_score = 1) then 1 else 0 end),2) as pct_nota1_d56,
       round(avg(case when (date_diff(tp.order_purchase_timestamp, '2018-07-01') <= 56) and (tor.review_score = 2) then 1 else 0 end),2) as pct_nota2_d56,

       round(avg(case when (date_diff(tp.order_purchase_timestamp, '2018-07-01') <= 365) and (tor.review_score = 1) then 1 else 0 end),2) as pct_nota1_d365,
       round(avg(case when (date_diff(tp.order_purchase_timestamp, '2018-07-01') <= 365) and (tor.review_score = 2) then 1 else 0 end),2) as pct_nota2_d365

from tb_pedidos as tp left join olist.order_reviews as tor on tp.order_id = tor.order_id
                      left join olist.order_items as toi on tp.order_id = toi.order_id

group by toi.seller_id 
```
