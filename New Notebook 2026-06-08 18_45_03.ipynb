```sql
-- -- Percentual de Notas Boas:
-- avaliações nota 4 ou 5 / quantidade de pedidos recebidos pelo seller

WITH base AS (
    SELECT DISTINCT
        oi.seller_id,
        oi.order_id,
        o.order_purchase_timestamp,
        r.review_score
    FROM workspace.olist.order_items oi
    JOIN workspace.olist.orders o
        ON oi.order_id = o.order_id
    LEFT JOIN workspace.olist.order_reviews r
        ON oi.order_id = r.order_id
    WHERE o.order_purchase_timestamp <= '2018-07-02'
)

SELECT
    seller_id,

    -- D-28
    SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02'
             AND review_score IN (4, 5)
            THEN 1 ELSE 0 
        END) * 1.0
    / NULLIF(SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02'
            THEN 1 ELSE 0 
        END), 0) AS pct_notas_boas_d28,

    -- D-56
    SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02'
             AND review_score IN (4, 5)
            THEN 1 ELSE 0 
        END) * 1.0
    / NULLIF(SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02'
            THEN 1 ELSE 0 
        END), 0) AS pct_notas_boas_d56,

    -- D-365
    SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02'
             AND review_score IN (4, 5)
            THEN 1 ELSE 0 
        END) * 1.0
    / NULLIF(SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02'
            THEN 1 ELSE 0 
        END), 0) AS pct_notas_boas_d365,

    -- Lifetime
    SUM(CASE 
            WHEN review_score IN (4, 5)
            THEN 1 ELSE 0 
        END) * 1.0
    / NULLIF(COUNT(order_id), 0) AS pct_notas_boas_lifetime

FROM base
GROUP BY seller_id;
```

```sql
-- +++++++++++++    PERCENTUAL NOTAS RUINS   +++++++++++++++++++++++++++++++++++++
-- Percentual de Notas Ruins:
-- avaliações nota 1 ou 2 / quantidade de pedidos recebidos pelo seller

WITH base AS (
    SELECT DISTINCT
        oi.seller_id,
        oi.order_id,
        o.order_purchase_timestamp,
        r.review_score
    FROM workspace.olist.order_items oi
    JOIN workspace.olist.orders o
        ON oi.order_id = o.order_id
    LEFT JOIN workspace.olist.order_reviews r
        ON oi.order_id = r.order_id
    WHERE o.order_purchase_timestamp <= '2018-07-02'
)

SELECT
    seller_id,

    -- D-28
    SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02'
             AND review_score IN (1, 2)
            THEN 1 ELSE 0 
        END) * 1.0
    / NULLIF(SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2018-06-04' AND '2018-07-02'
            THEN 1 ELSE 0 
        END), 0) AS pct_notas_ruins_d28,

    -- D-56
    SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02'
             AND review_score IN (1, 2)
            THEN 1 ELSE 0 
        END) * 1.0
    / NULLIF(SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2018-05-07' AND '2018-07-02'
            THEN 1 ELSE 0 
        END), 0) AS pct_notas_ruins_d56,

    -- D-365
    SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02'
             AND review_score IN (1, 2)
            THEN 1 ELSE 0 
        END) * 1.0
    / NULLIF(SUM(CASE 
            WHEN order_purchase_timestamp BETWEEN '2017-07-02' AND '2018-07-02'
            THEN 1 ELSE 0 
        END), 0) AS pct_notas_ruins_d365,

    -- Lifetime
    SUM(CASE 
            WHEN review_score IN (1, 2)
            THEN 1 ELSE 0 
        END) * 1.0
    / NULLIF(COUNT(order_id), 0) AS pct_notas_ruins_lifetime

FROM base
GROUP BY seller_id;
```
